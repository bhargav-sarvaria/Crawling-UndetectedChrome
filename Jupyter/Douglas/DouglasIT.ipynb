```python
import undetected_chromedriver.v2 as uc
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from importlib.machinery import SourceFileLoader
import time
from selenium import webdriver
import json
```

```python
driver = webdriver.Chrome('/Users/bhargavsarvaria/Desktop/Applications/seldoc/chrome/driver_mac/chromedriver')

base_url = 'https://www.douglas.it'
driver.get(base_url)
```

```python
def getFilename(conf):
    file_name = page_config['country'] + '_' + page_config['retailer'] + '_' + page_config['department'] + '_' + page_config['category'] + '_' + page_config['page']
    for k, v in json.load(open('/Users/bhargavsarvaria/Desktop/Applications/seldoc/config/encoding.json', 'r')).items():
        file_name = file_name.replace(k, v)
        file_name = file_name.replace('\\', '')
    return file_name

country = 'Spain'
code = 'es'

conf = {}
conf['country'] = country
conf['retailer'] = 'DOUGLAS_' + code.upper()
conf['parsing_config'] = './retailer_config/Douglas/douglas_parsing.json'
conf['gcloud_path'] = 'crawl_data/'+country+'/Douglas_' + code.upper()+'/'
page_configs = []
```

```python
conf
```

```python
departments = driver.find_elements(By.CLASS_NAME, 'rd__nav-main__item')

for department in departments:
    action = ActionChains(driver)
    action.move_to_element(department).perform()
    time.sleep(4)
    
    department_name = department.find_element(By.XPATH, './a').text
    department_url = department.find_element(By.XPATH, './a').get_attribute('href')
    print(department_name + ' | ' + department_name + ' | ' + department_name)
    print(department_url)
        
    page_config = conf.copy()            
    page_config['department'] = department_name
    page_config['category'] = department_name
    page_config['page'] = department_name
    page_config['page_url'] = department_url
    page_config['file_name'] = getFilename(page_config)
    page_configs.append(page_config)
    
        
    categories = driver.find_element(By.CLASS_NAME, 'rd__megadropdown-wrapper').find_elements(By.CLASS_NAME, 'rd__megadropdown__nav-block')
    for category in categories:
        if category.find_elements(By.XPATH, './a'):
            category_name = category.find_element(By.XPATH, './a').text
            category_link = category.find_element(By.XPATH, './a').get_attribute('href')
            
            print(department_name + ' | ' + category_name + ' | ' + category_name)
            print(category_link)

            page_config = conf.copy()            
            page_config['department'] = department_name
            page_config['category'] = category_name
            page_config['page'] = category_name
            page_config['page_url'] = category_link
            page_config['file_name'] = getFilename(page_config)
            page_configs.append(page_config)
        elif category.find_elements(By.CLASS_NAME, 'rd__headline'):
            category_name = category.find_element(By.CLASS_NAME, 'rd__headline').text
        elif category.find_elements(By.CLASS_NAME, 'rd__copytext'):
            category_name = category.find_element(By.CLASS_NAME, 'rd__copytext').text
            category_link = category.find_element(By.CLASS_NAME, 'rd__copytext').get_attribute('href')
            
            print(department_name + ' | ' + category_name + ' | ' + category_name)
            print(category_link)

            page_config = conf.copy()            
            page_config['department'] = department_name
            page_config['category'] = category_name
            page_config['page'] = category_name
            page_config['page_url'] = category_link
            page_config['file_name'] = getFilename(page_config)
            page_configs.append(page_config)
        
        pages = category.find_elements(By.CLASS_NAME, 'rd__list__item')
        for page in pages:
            page_name = page.find_element(By.TAG_NAME, 'a').text
            page_link = page.find_element(By.TAG_NAME, 'a').get_attribute('href')
            print(department_name + ' | ' + category_name + ' | ' + page_name)
            print(page_link)

            page_config = conf.copy()            
            page_config['department'] = department_name
            page_config['category'] = category_name
            page_config['page'] = page_name
            page_config['page_url'] = page_link
            page_config['file_name'] = getFilename(page_config)
            page_configs.append(page_config)
```
